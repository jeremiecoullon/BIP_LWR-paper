```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
# from matplotlib import cm
# import pandas as pd
```

```python

def FD_neg_power(rho, w, u, rho_j, z):
    return z * ( (u*rho/rho_j)**(-w) + (1-rho/rho_j)**(-w) )**(-1/w)
```

```python
Z, rho_j, u = 1, 1, 3.1
x_range = np.arange(0, rho_j, 0.001)

gamma_list = [0.5, 1, 5, 20, 100]

plt.rcParams.update({'font.size': 16})
plt.figure(figsize=(13, 9))

for gamma in gamma_list:
    FD_1 = {'z': Z, 'rho_j': rho_j, 'u': u, 'w': gamma}
    FD_fun_1 = lambda x: FD_neg_power(rho=x, **FD_1)
    FD_vals = list(map(FD_fun_1, x_range))
    plt.plot(x_range, FD_vals, label=fr"$\gamma={gamma}$")


plt.xlabel("Density", size=20)
plt.ylabel("Flow", size=20)


plt.legend(fontsize=20)
# plt.savefig("plots/delCast_FD-vary_w.png")
plt.show()
```
